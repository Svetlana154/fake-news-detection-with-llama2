```python
display(df_fa_kes['labels'].value_counts())
df_fa_kes['labels'].replace({0:False, 1:True}, inplace=True)
df_fa_kes['labels'].replace({False:1, True:0}, inplace=True)
display(df_fa_kes.head())
display(df_fa_kes['labels'].value_counts())
```

```python
df_fa_kes.rename(columns={'article_title':'title','article_content':'text', 'labels':'label'}, inplace=True)
display(df_fa_kes.head())
```

```python
df_fa_kes = df_fa_kes[['text', 'label', 'title', 'date', 'location', 'source']]
df_fa_kes.head()
```

```python
def remove_date(text):
  pattern = "(.* *\d{1,2} [a-zA-Z]* \d{4})|(.* [a-zA-Z]* *\d{1,2} *\d{4})|(\d{2}\.\d{2}\.\d{4}(\(.*\)){0,1})|(\d{2}-\d{2}-\d{4})"
  clean_text = re.sub(pattern, '', text).strip()
  return clean_text
```

```python
result = remove_date("Wed 05 Apr 2017 Syria attack symptoms consistent with")
result
```

```python
df_fa_kes['text'] = df_fa_kes['text'].apply(remove_date)
df_fa_kes.head()
```

```python
df_fa_kes = df_fa_kes[['text', 'label']].assign(metadata=df_fa_kes.iloc[:,2:].agg(dict,1))
display(df_fa_kes.head())
```
