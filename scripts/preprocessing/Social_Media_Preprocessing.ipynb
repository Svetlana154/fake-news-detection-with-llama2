```python
from datasets import load_dataset
import pandas as pd
def pd_set_see_full_dataframe():
    # Permanently changes the pandas settings
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    # pd.set_option('display.max_colwidth', -1)

def pd_reset_options():
    # Resets the options
    pd.reset_option('all')

def display_whole_df(df):
    pd_set_see_full_dataframe()
    display(df)
    pd_reset_options()
```

```python
df_gossip_false = pd.read_csv('/content/drive/MyDrive/CSI4900/Datasets/ForTraining/GossipCop/gossipcop_fake.csv')
df_gossip_true = pd.read_csv('/content/drive/MyDrive/CSI4900/Datasets/ForTraining/GossipCop/gossipcop_real.csv')
```

```python
df_gossip_true['label'] = 0
df_gossip_false['label'] = 1
```

```python
df_gossip = pd.concat([df_gossip_true, df_gossip_false], axis=0, ignore_index=True)
df_gossip.head()
```

```python
df_gossip['label'].value_counts()
```

```python
df_gossip.rename(columns={'title':'text', 'news_url':'link'}, inplace=True)
df_gossip = df_gossip[['text', 'label', 'link', 'tweet_ids']]
df_gossip.head()
```

```python
df_gossip = df_gossip[['text', 'label']].assign(metadata=df_gossip.iloc[:,2:].agg(dict,1))
df_gossip.head()
```

```python
print("Total number of records:", len(df_gossip))
print("Label distribution:")
print(df_gossip['label'].value_counts())
```
